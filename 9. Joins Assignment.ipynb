```
circuit_df = circuits_df.withColumnRenamed("location","circuit_location")
```

```
driver_df = driver_df.withColumnRenamed("full_name","driver_name")\
.withColumnRenamed("number", "driver_number")\
.withColumnRenamed("nationality","driver_nationality")
display(driver_df)
```

##Joining

```
circuit_races_df = races_df_final.join(circuit_df_filter, races_df_final.circuit_id == circuit_df_filter.circuit_id,"inner") \
.select(races_df_final.race_id, races_df_final.race_year ,races_df_final.race_name,races_df_final.race_timestamp, circuit_df_filter.circuit_location)
```

```
results_final_df = result_df.join(circuit_races_df, results_df.race_id == circuit_races_df.race_id) \
                              .join(driver_df, results_df.driver_id == driver_df.driver_id) \
                              .join(constructor_df, results_df.constructor_id == constructor_df.constructor_id)
display(results_final_df)
```

```
final_df = results_final_df.select("race_year","race_name","race_timestamp","circuit_location","driver_number",'driver_name',"driver_nationality","team","grid","fastest_lap","race_time","points") \
.withColumn("create_date", current_timestamp())
```

```
display(final_df.where("race_name = 'Abu Dhabi Grand Prix' and race_year = 2020") \
       .orderBy(final_df.points.desc()))
```
